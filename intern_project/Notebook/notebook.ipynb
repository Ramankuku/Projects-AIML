```python
import pandas as pd
data = pd.read_csv("C:\\Intern\\Data\\bnpl_user_dataset_realistic_labels.csv")
```

```python
df = data.copy()
```

```python
df.head()
```

```python
df.drop(columns=['age', 'orders_per_month', 'avg_order_amount', 'total_monthly_emi', 'dti_ratio', 'repayment_delay_count', 'last_payment_delay_days', 'on_time_repayment_rate','spending_volatility','impulse_purchase_count', 'risk_label', 'bnpl_utilization_ratio'
                 ], inplace=True)
```

```python
df[df['average_emi'] > df['monthly_income']]
```

```python
df = df[df['average_emi'] < df['monthly_income']]
```

```python
df['bnpl_income_ratio'] = df['average_emi']/df['monthly_income']
```

```python
df[(df['bnpl_income_ratio']>=0.3) & (df['bnpl_income_ratio']<=0.6)]
```

```python
def create_target(ratio):
    if ratio <=0.21:
        return 'Low Risk'
    elif 0.3 < ratio <= 0.6:
        return 'Medium Risk'
    else:
        return 'Risky'
```

```python
df
```

```python
df['Financial_Stress'] = df['bnpl_income_ratio'].map(create_target)
```

```python
rows_drop = df[df['num_active_bnpl_plans']<=1].sample(frac=0.31, random_state=42).index
df = df.drop(rows_drop)
```

```python
df.drop(columns=['user_id'], inplace=True) 
```

```python
df.insert(0, 'user_id', ['USR_' + str(i).zfill(5) for i in range(1, len(df) + 1)])
```

```python
df = df.drop(
    df[(df['average_emi'] == 0) & (df['total_bnpl_amount'] > 0)].index
)
```

```python
df 
```

```python
import string
import secrets
def generate_secure_password(length=12):
    """
    Generates a secure random password of a specified length.
    """
    alphabet = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(secrets.choice(alphabet) for i in range(length))
    
    return password

df['password'] = [generate_secure_password(length=12) for _ in range(len(df))]
```

```python
df[df['bnpl_income_ratio']>0.19]
```

```python
def create_deplayed(data):
    if data >0.19:
        return 1
    return 0

df['Emi_delayed'] = df['bnpl_income_ratio'].map(create_deplayed)
```

```python
df.drop(columns=['Financial_Stress']).to_csv('final_data.csv', index=False)
```
