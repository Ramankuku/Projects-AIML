# It shows 4 different values in Gender or I add the F-> Female and M->Male

```python
def preprocess_gender(data):
    if data == 'F':
        return 'Female'
    elif data == 'M':
        return 'Male'
    
    else:
        return data
    
    
```

```python
df['Gender'] = df['Gender'].apply(preprocess_gender)
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df[['Gender']] = imputer.fit_transform(df[['Gender']])
```
